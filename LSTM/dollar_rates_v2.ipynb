```python
import pandas as pd

# Load your CSV data into a DataFrame
df = pd.read_csv('/content/drive/MyDrive/Research/Dollar/daily dollar rate.csv', parse_dates=['DATE'])
df.head(40)
```

```python
# Replace '.' with NaN for missing values in 'DEXSLUS'
df['DEXSLUS'] = df['DEXSLUS'].replace('.', pd.NaT)

# Convert 'DEXSLUS' column to numeric
df['DEXSLUS'] = pd.to_numeric(df['DEXSLUS'], errors='coerce')

# Impute missing values with the previous available value in both columns
df['DEXSLUS'] = df['DEXSLUS'].fillna(method='ffill')

# Convert 'DATE' column to datetime
df['DATE'] = pd.to_datetime(df['DATE'])

df.head(40)
```

```python
import pandas as pd

# Assuming df is your DataFrame
# If 'date' is not in datetime format, convert it
df['DATE'] = pd.to_datetime(df['DATE'])

# Set the 'DATE' column as the index
df = df.set_index('DATE')

# Resample the data to fill in missing dates
df_resampled = df.resample('D').interpolate(method='linear')

# Reset the index to bring 'DATE' back as a column
df = df_resampled.reset_index()

# Display the result
print(df)


```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam

# Load your dataset
# Assuming 'date' is in datetime format and 'DEXSLUS' is your exchange rate
# df = ...

# Use MinMaxScaler to normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
df['DEXSLUS_normalized'] = scaler.fit_transform(df['DEXSLUS'].values.reshape(-1, 1))

# Function to prepare the data for LSTM
def create_lstm_dataset(dataset, time_steps=1):
    dataX, dataY = [], []
    for i in range(len(dataset) - time_steps):
        a = dataset[i:(i + time_steps), 0]
        dataX.append(a)
        dataY.append(dataset[i + time_steps, 0])
    return np.array(dataX), np.array(dataY)

# Set the time steps (you can adjust this based on your dataset)
time_steps = 3

# Create the LSTM dataset
X, y = create_lstm_dataset(df[['DEXSLUS_normalized']].values, time_steps)

# Reshape the input data for LSTM
X = X.reshape(X.shape[0], X.shape[1], 1)

# Split the data into training, validation, and testing sets
split_train = int(0.8 * len(X))
split_val = int(0.85 * len(X))

X_train, X_val, X_test = X[:split_train], X[split_train:split_val], X[split_val:]
y_train, y_val, y_test = y[:split_train], y[split_train:split_val], y[split_val:]

# Build the LSTM model with provided hyperparameters
model = Sequential()
model.add(LSTM(40, input_shape=(X_train.shape[1], 1), return_sequences=True, activation='tanh'))
model.add(Dropout(0.2))
model.add(LSTM(30, return_sequences=True, activation='tanh'))
model.add(Dropout(0.2))
model.add(LSTM(10, activation='tanh'))
model.add(Dense(1))

adam_optimizer = Adam(learning_rate=0.0005)
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])

# Use EarlyStopping to stop training if the validation loss doesn't improve
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
# Train the model with validation data
history = model.fit(X_train, y_train, epochs=30, batch_size=50, verbose=2, validation_data=(X_val, y_val), callbacks=[early_stopping])

# Make predictions on the test set
y_pred = model.predict(X_test)

# Invert the normalization to get the actual values
y_pred_actual = scaler.inverse_transform(y_pred)
y_test_actual = scaler.inverse_transform(y_test.reshape(-1, 1))

# Calculate Mean Absolute Error
mae = mean_absolute_error(y_test_actual, y_pred_actual)
print(f'Mean Absolute Error: {mae}')
```
